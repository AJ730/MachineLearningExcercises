```python
%matplotlib widget
W_perceptron = [0, 0, 0]
X_new = np.c_[X, np.ones(X.shape[0])] 
# from sklearn.linear_model import Perceptron

def train_perceptron(X, labels, init_w, rho):
    #Check if w.Tx < or >= 0
    correct = 0
    w = init_w
    while correct != len(labels):
        correct = 0
        for i in range(len(X)):
            current = X[i]
            if(np.dot(w, current) <0 and labels[i] == -1):
                correct += 1
            elif(np.dot(w, current) >0 and labels[i] == 1):
                correct += 1
            else:
                w = w + labels[i]*rho*current            
    return w


```
